Read CSV using NumPy

```python
import numpy as np

data = np.genfromtxt("/content/StudentsPerformance.csv", delimiter=",", dtype=str)
data
```

Separate header and rows


```python
header = data[0]   # column names
rows = data[1:]    # student records
```

Remove duplicate rows (NumPy)

```python
unique_rows = np.unique(rows, axis=0)
```

Combine header + cleaned rows

```python
clean_data = np.vstack((header, unique_rows))
```

Save cleaned CSV file

```python
np.savetxt("cleaned_students.csv",
           clean_data,
           delimiter=",",
           fmt="%s")
```
