```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Students.csv.zip")
print(df.head())
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print(df.columns)
```

Bar Chart — Average Marks per Subject

```python
subjects_avg = df[["math score","reading score","writing score"]].mean()

subjects_avg.plot(kind="bar")
plt.title("Average Marks in Each Subject")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.show()
```

Pie Chart — Total Marks Distribution

```python
total_marks = df[["math score","reading score","writing score"]].sum()

plt.pie(total_marks, labels=total_marks.index, autopct='%1.1f%%')
plt.title("Overall Marks Distribution")
plt.show()
```

Histogram — Math Marks Distribution

```python
plt.hist(df["math score"], bins=10)
plt.title("Distribution of Math Marks")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```

Line Chart — Student Performance Comparison


```python
df[["math score","reading score","writing score"]].plot(marker='o')
plt.title("Marks Comparison of Students")
plt.xlabel("Student Index")
plt.ylabel("Marks")
plt.show()
```

Box Plot — Performance Spread

```python
df[["math score","reading score","writing score"]].plot(kind="box")
plt.title("Marks Spread (Min–Max & Median)")
plt.show()
```

Scatter Plot — Study Hours vs Math Marks

```python
plt.scatter(df["reading score"], df["math score"])
plt.title("Reading Score vs Math Marks")
plt.xlabel("Reading Score")
plt.ylabel("Math Marks")
plt.show()
```

Attendance Impact Chart

```python
plt.scatter(df["reading score"], df["writing score"])
plt.title("Reading Score vs Writing Score")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.show()
```
